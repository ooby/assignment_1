```python
import numpy as np
import matplotlib.pyplot as plt

def plot_3d_matrix(d):
    fig = plt.figure(figsize=(10, 10), dpi=100)
    ax = fig.add_subplot(111, projection='3d')
    z, x, y = d.nonzero()
    ax.scatter(x, y, z, c = z, alpha = 1)
    plt.show()
```

```python
shape = 20 # Размеры матрицы

i = np.identity(shape, dtype=np.uint8)
data = np.repeat(i, shape, axis = 0).reshape((shape, shape, shape))
print(data.shape)
```

```python
plot_3d_matrix(data)
```

```python
i = np.identity(shape, dtype=np.uint8)
test = np.repeat(i, shape, axis = 1).reshape((shape, shape, shape))
print(test.shape)
```

```python
plot_3d_matrix(test)
```

1. Дана 3х-мерная матрица `data` с 1 по условию z = y и 0 по всем другим условиям (картинка 1). Необходимо разработать функцию `rotate`, которая поворачивает данную матрицу так, чтобы она совпала с 3х-мерной матрицей `test`.
2. Необходимо сделать проверку `assert test == rotate(data)`
